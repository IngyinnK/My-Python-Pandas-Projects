## loc and iloc

```python
# name of the student at first row
df.loc[0, "name"]
# 0 is the index number for the first row.
```

```python
df.iloc[0, 1]
# 0 is the index number for the first row.
# 1 is the index number for "name" column.
```

```python
# first 5 rows, id, name, gender
df.loc[0:4, ["id", "name", "gender"]]
# 0:4 is first 5 rows.
# these are list of the column names  ["id", "name", "gender"].
# loc operator includes stopping point of index.
```

```python
df.iloc[0:5, [0, 1, 4]]
# iloc operator does not include stopping point of index.
```

```python
df.loc[27, "class"]
```

```python
df.loc[:4, "name"]
# :4 starts from first index, print until index 4
```

```python
df.loc[29: , "class"]
# 29: start from index 29 till the last
```

## How to filter (you cannot use iloc in filtering)

```python
df["mark"]>60
# boolen masking (true or false)
```

```python
df[df["mark"]>60]
```

```python
# df[condition]
```

```python
df[df["class"]=="Six"]
```

```python
# df[] slicing method
# df["class"]=="Six" filtering rows
```

```python
df[df["gender"]=="female"].head()
```

```python
df[df["mark"]>90]
```

```python
# female student at class four
# slicing method for multiple conditions >> df[(condition1) & (condition2)]
df[(df["gender"]=="female") & (df["class"]=="Four")]
```

```python
# create report of student with their mark less than 70 at class seven
df[(df["mark"]<70) & (df["class"]=="Seven")]
```

df.loc[condition, column name(s)]

```python
# df[df["gender"]=="female"].head()
# df.loc[rows, coloumn names]

df.loc[df["gender"]=="female", :].head()

# : return all columns
```

```python
df.loc[df["gender"]=="female", ["name", "gender"]].head()
```

```python
# df[df["mark"]>90]

df.loc[df["mark"]>90, :]
```

- if you are using slicing method for a single column df[""]

- ["coloumn1", "coloumn2"] is for list of columnn names
- filter data for single column df[df["mark"]>90]
- filter data for multipe column df[(condition1) & (condition2)]

- if you are using df.loc[] or df.iloc[]
- if you are using for filtering df.loc[df["mark">90], :]

```python
df.loc[0:4, ["name", "mark"]].sort_values(by="mark", ascending=False)
# ascending=False will sot in descending order
```

```python
# find male student with their mark higher than 70 and sort by their class
df.loc[(df["gender"]=="male") & (df["mark"]>70)].sort_values(by="class")
```

```python
# Write code to find information of female students who are above the classroom average. Sort the results in descening order.
df[(df["gender"]=="female") & (df["mark"]>df["mark"].mean())].sort_values(by="mark", ascending=False)
```
